```python
#计算净收益率从-20%到20%时两者差距
import numpy as np
import pandas as pd
R=np.arange(-0.2,+0.21,0.02)
r=np.log(1+R)#对于每个element进行log
R_r_diff=R-r
res=[R,r,R_r_diff]
res=pd.DataFrame(res,index=['SimpleRet','LogRet','Diff'],columns=R.round(2))#对R中的每个元素进行round(2)，两位小数round
print(res[res.columns[np.arange(0,21,4)]])#每隔四列取出来看
```
